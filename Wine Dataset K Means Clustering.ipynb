```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = 'Wine dataset.csv'
df = pd.read_csv('Wine dataset.csv')

# Check the first few rows to understand the data
print(df.head())

# Extract features and labels
features = df.drop('class', axis=1)
labels = df['class']

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Add cluster labels to the dataframe
df['cluster'] = clusters

# Plot the clusters with actual labels for comparison
plt.figure(figsize=(12, 6))
sns.scatterplot(x=scaled_features[:, 0], y=scaled_features[:, 1], hue=labels, style=clusters, palette='Set1')
plt.title('KMeans Clusters vs Actual Labels')
plt.xlabel('Standardized Alcohol')
plt.ylabel('Standardized Malic acid')
plt.legend(title='Class / Cluster')
plt.show()

# Show the first few rows with cluster labels
print(df.head())
```
